## Torres de Hanoi con Divide y vencerás

````python
def Torres_Hanoi(N: int, desde: int, hasta: int) -> None:
    if N == 1:
        print(f"Lleva la ficha {N} desde '{desde}' hasta '{hasta}'")
    else:
        Torres_Hanoi(N-1, desde, 6 - desde - hasta)
        print(f"Lleva la ficha {N} desde '{desde}' hasta '{hasta}'")
        Torres_Hanoi(N-1, 6 - desde - hasta, hasta)

Torres_Hanoi(3, 1, 3)
````

#### Versión adaptada de Torres de Hanoi para difeerentes tamaños

El código anterior usa `6 - desde - hasta` para calcular el poste auxiliar, lo cual solo funciona cuando los postes son exactamente 1, 2 y 3 (porque 1+2+3 = 6).

A continuación, una versión generalizada que recibe el poste auxiliar como parámetro, de forma que:
- Funciona con **cualquier número de fichas**
- Los postes pueden llamarse como queramos (letras, números, nombres...)
- Es más fácil de entender la lógica del algoritmo

````python
def Torres_Hanoi_General(N: int, desde: str, hasta: str, auxiliar: str) -> None:
    """
    Torres de Hanoi generalizada.
    - N: número de fichas a mover
    - desde: poste origen
    - hasta: poste destino
    - auxiliar: poste auxiliar (intermedio)

    Funciona con cualquier nombre de poste y cualquier número de fichas.
    """
    if N == 1:
        print(f"Lleva la ficha {N} desde '{desde}' hasta '{hasta}'")
    else:
        # 1. Mover N-1 fichas del origen al auxiliar (usando destino como puente)
        Torres_Hanoi_General(N - 1, desde, auxiliar, hasta)
        # 2. Mover la ficha grande del origen al destino
        print(f"Lleva la ficha {N} desde '{desde}' hasta '{hasta}'")
        # 3. Mover las N-1 fichas del auxiliar al destino (usando origen como puente)
        Torres_Hanoi_General(N - 1, auxiliar, hasta, desde)


# Ejemplo 1: Con 3 fichas (mismo resultado que el código anterior)
print("=== 3 fichas (postes 1, 2, 3) ===")
Torres_Hanoi_General(3, "1", "3", "2")

# Ejemplo 2: Con 4 fichas
print("\n=== 4 fichas (postes A, B, C) ===")
Torres_Hanoi_General(4, "A", "C", "B")

# Ejemplo 3: Con 5 fichas
print("\n=== 5 fichas (postes A, B, C) ===")
Torres_Hanoi_General(5, "A", "C", "B")
````

## Sucesión de Fibonacci

````python
# Sucesión_de_Fibonacci
# https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci
# Calculo del termino n-simo de la suscesión de Fibonacci
def Fibonacci(N: int) -> int:
    if N < 2:
        return 1
    else:
        return Fibonacci(N-1)+Fibonacci(N-2)

Fibonacci(5)
````

#### Análisis de complejidad y optimizaciones

El código recursivo anterior tiene **complejidad O(2ⁿ)** (exponencial), porque cada llamada genera dos llamadas más y se recalculan los mismos valores muchas veces. Por ejemplo, para `Fibonacci(5)`:

```
                    F(5)
                /         \
            F(4)            F(3)       ← F(3) se calcula 2 veces
           /    \          /    \
        F(3)    F(2)    F(2)    F(1)   ← F(2) se calcula 3 veces
        /  \
     F(2) F(1)
```

A continuación, tres técnicas para reducir la complejidad:

| Método | Tiempo | Espacio | Idea clave |
|---|---|---|---|
| Recursivo original | O(2ⁿ) | O(n) | Recalcula todo cada vez |
| Memoización | O(n) | O(n) | Guarda resultados en diccionario |
| Prog. dinámica | O(n) | O(n) | Construye de abajo a arriba |
| Prog. dinámica optimizada | O(n) | **O(1)** | Solo guarda los 2 últimos valores |

````python
import time

# ============================================================
# 1) MEMOIZACIÓN - O(n) tiempo, O(n) espacio
# Guardamos en un diccionario los valores ya calculados
# para no repetir trabajo
# @more: https://es.wikipedia.org/wiki/Memoizaci%C3%B3n
# ============================================================
def Fibonacci_memo(N: int, memo: dict = {}) -> int:
    if N in memo:
        return memo[N] # Ya lo calculamos antes
    if N < 2:
        return 1
    else:
        memo[N] = Fibonacci_memo(N - 1, memo) + Fibonacci_memo(N - 2, memo)
        return memo[N]


# ============================================================
# 2) PROGRAMACIÓN DINÁMICA - O(n) tiempo, O(n) espacio
# Construimos la solución de abajo a arriba con una lista
# Sin recursión -> sin riesgo de desbordamiento de pila
# ============================================================
def Fibonacci_pd(N: int) -> int:
    if N < 2:
        return 1
    tabla = [0] * (N + 1)
    tabla[0] = 1 # F(0) = 1
    tabla[1] = 1 # F(1) = 1
    for i in range(2, N + 1):
        tabla[i] = tabla[i - 1] + tabla[i - 2]
    return tabla[N]


# ============================================================
# 3) PROGRAMACIÓN DINÁMICA OPTIMIZADA - O(n) tiempo, O(1) espacio
# Solo necesitamos los 2 valores anteriores, no toda la lista
# ============================================================
def Fibonacci_optimo(N: int) -> int:
    if N < 2:
        return 1
    anterior = 1 # F(0)
    actual = 1 # F(1)
    for i in range(2, N + 1):
        anterior, actual = actual, anterior + actual
    return actual


# ============================================================
# COMPARACIÓN DE LAS 4 VERSIONES
# ============================================================
N_test = 30 # Usamos N=30 para que se note la diferencia

print(f"Calculando Fibonacci({N_test}) con cada método:\n")

# Original recursivo (O(2^n)) - ¡CUIDADO: con N>35 puede tardar mucho!
inicio = time.time()
resultado = Fibonacci(N_test)
t_recursivo = time.time() - inicio
print(f"  Recursivo original: {resultado:>12}   tiempo: {t_recursivo:.6f}s   complejidad: O(2^n)")

# Memoización (O(n))
Fibonacci_memo_clean = {}  # Limpiamos la caché para medir bien
def Fibonacci_memo_test(N, memo=Fibonacci_memo_clean):
    if N in memo:
        return memo[N]
    if N < 2:
        return 1
    memo[N] = Fibonacci_memo_test(N - 1, memo) + Fibonacci_memo_test(N - 2, memo)
    return memo[N]

inicio = time.time()
resultado = Fibonacci_memo_test(N_test)
t_memo = time.time() - inicio
print(f"  Memoización:        {resultado:>12}   tiempo: {t_memo:.6f}s   complejidad: O(n)")

# Programación dinámica (O(n))
inicio = time.time()
resultado = Fibonacci_pd(N_test)
t_pd = time.time() - inicio
print(f"  Prog. dinámica:     {resultado:>12}   tiempo: {t_pd:.6f}s   complejidad: O(n)")

# PD optimizada (O(n) tiempo, O(1) espacio)
inicio = time.time()
resultado = Fibonacci_optimo(N_test)
t_optimo = time.time() - inicio
print(f"  PD optimizada:      {resultado:>12}   tiempo: {t_optimo:.6f}s   complejidad: O(n), espacio O(1)")

# Speedup
if t_memo > 0:
    print(f"\n  Speedup recursivo vs memoización: {t_recursivo / t_memo:.0f}x")
if t_pd > 0:
    print(f"  Speedup recursivo vs programación dinámica: {t_recursivo / t_pd:.0f}x")
if t_optimo > 0:
    print(f"  Speedup recursivo vs programación dinámica optmizada: {t_recursivo / t_optimo:.0f}x")
````

## Devolución de cambio por técnica voraz

````python
def cambio_monedas(N: int, SM: list):
    SOLUCION = [0] * len(SM) #SOLUCION = [0,0,0,0,..]
    ValorAcumulado = 0

    for i, valor in enumerate(SM):
        monedas = (N - ValorAcumulado) // valor
        SOLUCION[i] = monedas
        ValorAcumulado = ValorAcumulado + (monedas * valor)

        if ValorAcumulado == N:
            return SOLUCION

print(cambio_monedas(15,[25,10,5,1]))
print(cambio_monedas(15,[10,25,1, 5]))
````

## N-Reinas por técnica de vueta atrás


````python
def escribe(S: list) -> None:
    n = len(S)
    for x in range(n):
        print("")
        for i in range(n):
            if S[i] == x+1:
                print(" X " , end="")
            else:
                print(" - ", end="")

def es_prometedora(SOLUCION: list, etapa: int, show_trace: bool = False):
    if show_trace:
        print("\n"+"*"*20, end="\n")
        print(SOLUCION)
        escribe(SOLUCION)
        print("\n"+"*"*20, end="\n")

    # Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
    for i in range(etapa+1):
        #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
        if SOLUCION.count(SOLUCION[i]) > 1:
            return False

        #Verifica las diagonales
        for j in range(i+1, etapa +1 ):
            #print("Comprobando diagonal de " + str(i) + " y " + str(j))
            if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
    return True

def reinas(N, solucion=[], etapa=0, show_trace: bool = False):
    if len(solucion) == 0:
        solucion=[0 for i in range(N)]

    for i in range(1, N+1):
        solucion[etapa] = i

        if es_prometedora(solucion, etapa, show_trace=show_trace):
            if etapa == N-1:
                print(solucion)
                #escribe(solucion)
                print()
            else:
                reinas(N, solucion, etapa+1)
        solucion[etapa] = 0

# reinas(4, show_trace=True)
reinas(4, show_trace=False)
````

````python
# NOTA: Esto sería la Matriz de adyacencia
# @more: https://es.wikipedia.org/wiki/Matriz_de_adyacencia
# Si miramos la diagonal de la matriz, todos son cero, porque sería ir del punto de origen al punto de destino
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]

################################################################
def Precios(TARIFAS):
################################################################
    #Total de Nodos
    N = len(TARIFAS[0])

    #Inicialización de la tabla de precios
    PRECIOS = [ [9999]*N for i in [9999]*N]
    RUTA = [ [""]*N for i in [""]*N]

    for i in range(0,N-1):
        RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
        PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0

        for j in range(i+1, N):
            MIN = TARIFAS[i][j]
            RUTA[i][j] = i

            for k in range(i, j):
                if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
                    MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
                    RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
                PRECIOS[i][j] = MIN

    return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
    print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
    print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
    if desde == hasta:
        #print("Ir a :" + str(desde))
        return ""
    else:
        return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                    ',' + \
                    str(RUTA[desde][hasta] \
                )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0, 6)
````

## Práctica Individual

**Problema: Dos puntos más cercanos**
- Dado un conjunto de puntos, encontrar los dos puntos más cercanos entre sí.
- Guía para aprendizaje:
    - Suponer en 1D, o sea, una lista de números: [3403, 4537, 9089, 9746, 7259, ...]
    - Primer intento: Fuerza bruta
        - Calcular la complejidad. ¿Se puede mejorar?
    - Segundo intento. Aplicar Divide y Vencerás
        - Calcular la complejidad. ¿Se puede mejorar?
    - Extender el algoritmo a 2D: [(1122, 6175), (135, 4076), (7296, 2741)…]
    - Extender el algortimo a 3D.

### Generación de datos aleatorios

Generamos conjuntos de datos aleatorios para probar los algoritmos en 1D, 2D y 3D.
- **1D:** Lista de 1000 números enteros aleatorios entre 1 y 10000
- **2D:** Lista de 1000 tuplas (x, y) con valores entre 1 y 10000
- **3D:** Lista de 1000 tuplas (x, y, z) con valores entre 1 y 10000

````python
import random
import time
import math

# Fijamos semilla para reproducibilidad
random.seed(5)

"""
Generación de datos aleatorios
NOTA: después de las pruebas se ha comprado que este código genera elementos repetidos:
LISTA_1D = [random.randrange(1, 10000) for x in range(1000)]

Por lo que se ha modificado por el método "sample":
LISTA_1D = random.sample(range(1, 10000), 1000)
"""
LISTA_1D = random.sample(range(1, 10000), 1000)
LISTA_2D = [(random.randrange(1, 10000), random.randrange(1, 10000)) for x in range(1000)]
LISTA_3D = [(random.randrange(1, 10000), random.randrange(1, 10000), random.randrange(1, 10000)) for x in range(1000)]

print(f"LISTA_1D: {len(LISTA_1D)} elementos. Ejemplo: {LISTA_1D[:5]}")
print(f"LISTA_2D: {len(LISTA_2D)} elementos. Ejemplo: {LISTA_2D[:5]}")
print(f"LISTA_3D: {len(LISTA_3D)} elementos. Ejemplo: {LISTA_3D[:5]}")
````

### 1D - Fuerza Bruta

**Idea:** Comparar cada punto con todos los demás y quedarnos con la pareja que tenga menor distancia.

**Complejidad:**
- Tenemos dos bucles anidados: el primero recorre n elementos, el segundo recorre los restantes.
- Total de comparaciones: n*(n-1)/2
- **Complejidad temporal: O(n²)**
- **Complejidad espacial: O(1)** (solo guardamos la mejor pareja)

En 1D la "distancia" entre dos puntos es simplemente el valor absoluto de su diferencia: |a - b|

````python
def puntos_cercanos_1d_fuerza_bruta(lista: list):
    """
    Encuentra los dos puntos más cercanos en una lista 1D usando fuerza bruta.
    Devuelve la distancia mínima y los dos puntos.
    """
    n = len(lista)
    min_distancia = float('inf')
    punto_a = None
    punto_b = None

    # Comparamos cada par de puntos
    for i in range(n):
        for j in range(i + 1, n):
            distancia = abs(lista[i] - lista[j])
            if distancia < min_distancia:
                min_distancia = distancia
                punto_a = lista[i]
                punto_b = lista[j]

    return min_distancia, punto_a, punto_b


# Ejecutamos y medimos el tiempo
inicio = time.time()
distancia, p1, p2 = puntos_cercanos_1d_fuerza_bruta(LISTA_1D)
fin = time.time()

print(f"Fuerza Bruta 1D:")
print(f"  Puntos más cercanos: {p1} y {p2}")
print(f"  Distancia: {distancia}")
print(f"  Tiempo: {fin - inicio:.6f} segundos")
````

### 1D - Divide y Vencerás

**Idea:** En 1D podemos mejorar la fuerza bruta con un truco sencillo:
1. **Ordenar** la lista de puntos (coste O(n log n))
2. Una vez ordenada, los dos puntos más cercanos siempre serán **vecinos** en la lista ordenada
3. Recorrer la lista comparando cada elemento con el siguiente (coste O(n))

**¿Por qué funciona?** Si tenemos tres números a ≤ b ≤ c, la distancia entre a y c siempre será mayor o igual que la distancia entre a y b o entre b y c. Por tanto, el par más cercano siempre son elementos consecutivos en la lista ordenada.

**Complejidad:**
- Ordenar: O(n log n)
- Recorrer vecinos: O(n)
- **Complejidad total: O(n log n)** — Mejora respecto a O(n²)
- **Complejidad espacial: O(n)** (por la copia ordenada)

> **Nota:** Técnicamente esto no es un Divide y Vencerás "puro" en el sentido recursivo, pero se basa en la misma idea: al ordenar estamos aprovechando la estructura del problema para reducir el espacio de búsqueda. El paso de "ordenar" se implementa internamente con algoritmos DyV como MergeSort.

````python
def puntos_cercanos_1d_dyv(lista: list):
    """
    Encuentra los dos puntos más cercanos en 1D usando ordenación (Divide y Vencerás).
    1. Ordenamos la lista
    2. Comparamos solo elementos consecutivos
    """
    # Ordenar la lista (internamente usa TimSort, basado en MergeSort -> DyV)
    lista_ordenada = sorted(lista)

    min_distancia = float('inf')
    punto_a = None
    punto_b = None

    # Solo comparamos vecinos consecutivos
    for i in range(len(lista_ordenada) - 1):
        distancia = abs(lista_ordenada[i] - lista_ordenada[i + 1])
        if distancia < min_distancia:
            min_distancia = distancia
            punto_a = lista_ordenada[i]
            punto_b = lista_ordenada[i + 1]

    return min_distancia, punto_a, punto_b


# Ejecutamos y medimos el tiempo
inicio = time.time()
distancia, p1, p2 = puntos_cercanos_1d_dyv(LISTA_1D)
fin = time.time()

print(f"Divide y Vencerás 1D:")
print(f"  Puntos más cercanos: {p1} y {p2}")
print(f"  Distancia: {distancia}")
print(f"  Tiempo: {fin - inicio:.6f} segundos")
````

### 2D - Fuerza Bruta

**Idea:** Igual que en 1D, pero ahora la distancia entre dos puntos (x1, y1) y (x2, y2) es la **distancia euclidiana**:

$$d = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}$$

Comparamos todos los pares posibles.

**Complejidad:**
- **Complejidad temporal: O(n²)** — Misma estructura de dos bucles anidados
- **Complejidad espacial: O(1)**

````python
def distancia_2d(p1: int, p2: int) -> float:
    """Calcula la distancia euclidiana entre dos puntos 2D."""
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
````

````python
def puntos_cercanos_2d_fuerza_bruta(lista: list):
    """
    Encuentra los dos puntos más cercanos en 2D usando fuerza bruta.
    Compara todos los pares posibles.
    """
    n = len(lista)
    min_distancia = float('inf')
    punto_a = None
    punto_b = None

    for i in range(n):
        for j in range(i + 1, n):
            d = distancia_2d(lista[i], lista[j])
            if d < min_distancia:
                min_distancia = d
                punto_a = lista[i]
                punto_b = lista[j]

    return min_distancia, punto_a, punto_b


# Ejecutamos y medimos el tiempo
inicio = time.time()
distancia, p1, p2 = puntos_cercanos_2d_fuerza_bruta(LISTA_2D)
fin = time.time()

print(f"Fuerza Bruta 2D:")
print(f"  Puntos más cercanos: {p1} y {p2}")
print(f"  Distancia: {distancia:.4f}")
print(f"  Tiempo: {fin - inicio:.6f} segundos")
````

### 2D - Divide y Vencerás

**Idea:** Este es el algoritmo clásico de Shamos & Hoey (1975), descrito en Cormen et al. Cap. 33.4:

1. **Ordenar** los puntos por coordenada X
2. **Dividir** el conjunto en dos mitades por la mediana de X
3. **Conquistar** recursivamente: encontrar el par más cercano en cada mitad
4. **Combinar:** El paso clave. Sea δ = min(distancia_izquierda, distancia_derecha). Los dos puntos más cercanos podrían estar uno en cada mitad, pero solo si ambos están a distancia < δ de la línea divisoria.
   - Crear una "franja" con los puntos a distancia < δ de la línea media
   - Ordenar la franja por coordenada Y
   - Para cada punto en la franja, solo necesitamos comparar con los siguientes 7 puntos (demostrado geométricamente)

**Complejidad:**
- Ordenar por X: O(n log n) (se hace una sola vez)
- Recurrencia: T(n) = 2·T(n/2) + O(n log n) para la franja
- **Complejidad total: O(n log² n)**
- Se puede optimizar a O(n log n) preordenando por Y, pero usamos la versión más sencilla.

> **Ref:** Shamos, M. I. & Hoey, D. (1975). "Closest-point problems". *16th Annual Symposium on Foundations of Computer Science*, IEEE.

````python
def puntos_cercanos_2d_dyv(lista: list):
    """
    Encuentra los dos puntos más cercanos en 2D usando Divide y Vencerás.
    Algoritmo de Shamos & Hoey.
    """

    def _fuerza_bruta_pequena(puntos: list) -> tuple:
        """Fuerza bruta para conjuntos pequeños (caso base)."""
        min_dist = float('inf')
        mejor_par = (None, None)
        n = len(puntos)
        for i in range(n):
            for j in range(i + 1, n):
                d = distancia_2d(puntos[i], puntos[j])
                if d < min_dist:
                    min_dist = d
                    mejor_par = (puntos[i], puntos[j])

        return min_dist, mejor_par

    def _franja_cercanos(franja: list, delta: float) -> tuple:
        """Busca el par más cercano en la franja central."""
        min_dist = delta
        mejor_par = (None, None)

        # Ordenar la franja por coordenada Y
        franja.sort(key=lambda p: p[1])

        # Para cada punto, comparar solo con los siguientes
        # (máximo 7 por la propiedad geométrica)
        for i in range(len(franja)):
            j = i + 1
            while j < len(franja) and (franja[j][1] - franja[i][1]) < min_dist:
                d = distancia_2d(franja[i], franja[j])
                if d < min_dist:
                    min_dist = d
                    mejor_par = (franja[i], franja[j])
                j += 1

        return min_dist, mejor_par

    def _dyv_recursivo(puntos_x: list) -> tuple:
        """Función recursiva principal."""
        n = len(puntos_x)

        # Caso base: si hay 3 o menos puntos, usar fuerza bruta
        if n <= 3:
            return _fuerza_bruta_pequena(puntos_x)

        # Dividir por la mediana de X
        mitad = n // 2
        punto_medio = puntos_x[mitad]

        izquierda = puntos_x[:mitad]
        derecha = puntos_x[mitad:]

        # Conquistar: resolver recursivamente cada mitad
        dist_izq, par_izq = _dyv_recursivo(izquierda)
        dist_der, par_der = _dyv_recursivo(derecha)

        # Tomar el mínimo de ambas mitades
        if dist_izq < dist_der:
            delta = dist_izq
            mejor_par = par_izq
        else:
            delta = dist_der
            mejor_par = par_der

        # Combinar: buscar en la franja central
        franja = []
        for p in puntos_x:
            if abs(p[0] - punto_medio[0]) < delta:
                franja.append(p)

        dist_franja, par_franja = _franja_cercanos(franja, delta)

        if dist_franja < delta:
            return dist_franja, par_franja
        else:
            return delta, mejor_par

    # Ordenar por X una sola vez
    puntos_ordenados = sorted(lista, key=lambda p: p[0])
    dist, par = _dyv_recursivo(puntos_ordenados)
    return dist, par[0], par[1]


# Ejecutamos y medimos el tiempo
inicio = time.time()
distancia, p1, p2 = puntos_cercanos_2d_dyv(LISTA_2D)
fin = time.time()

print(f"Divide y Vencerás 2D:")
print(f"  Puntos más cercanos: {p1} y {p2}")
print(f"  Distancia: {distancia:.4f}")
print(f"  Tiempo: {fin - inicio:.6f} segundos")
````

### 3D - Fuerza Bruta

**Extensión a 3D:** La distancia euclidiana en 3D entre (x1, y1, z1) y (x2, y2, z2) es:

$$d = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}$$

El enfoque de fuerza bruta es idéntico: comparar todos los pares.

**Complejidad:**
- **Complejidad temporal: O(n²)**
- **Complejidad espacial: O(1)**

````python
def distancia_3d(p1: int, p2: int) -> float:
    """Calcula la distancia euclidiana entre dos puntos 3D."""
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2 + (p1[2] - p2[2])**2)
````

````python
def puntos_cercanos_3d_fuerza_bruta(lista: list) -> tuple:
    """
    Encuentra los dos puntos más cercanos en 3D usando fuerza bruta.
    """
    n = len(lista)
    min_distancia = float('inf')
    punto_a = None
    punto_b = None

    for i in range(n):
        for j in range(i + 1, n):
            d = distancia_3d(lista[i], lista[j])
            if d < min_distancia:
                min_distancia = d
                punto_a = lista[i]
                punto_b = lista[j]

    return min_distancia, punto_a, punto_b


# Ejecutamos y medimos el tiempo
inicio = time.time()
distancia, p1, p2 = puntos_cercanos_3d_fuerza_bruta(LISTA_3D)
fin = time.time()

print(f"Fuerza Bruta 3D:")
print(f"  Puntos más cercanos: {p1} y {p2}")
print(f"  Distancia: {distancia:.4f}")
print(f"  Tiempo: {fin - inicio:.6f} segundos")
````

### 3D - Divide y Vencerás

**Idea:** La extensión del algoritmo DyV a 3D sigue la misma lógica que en 2D:

1. **Ordenar** por coordenada X
2. **Dividir** en dos mitades
3. **Conquistar** recursivamente
4. **Combinar:** En la franja, ahora debemos considerar puntos cercanos al plano divisor. Se crea un "slab" (losa) de ancho 2δ centrado en el plano X = mediana, y dentro de ese slab se ordenan los puntos por Y y se buscan pares cercanos.

**Complejidad:**
- **Complejidad total: O(n log² n)** — misma que en 2D con esta implementación

> **Ref:** El algoritmo generaliza a d dimensiones con complejidad O(n log^(d-1) n) según Preparata & Shamos (1985).

````python
def puntos_cercanos_3d_dyv(lista: list):
    """
    Encuentra los dos puntos más cercanos en 3D usando Divide y Vencerás.
    Extensión del algoritmo 2D al espacio tridimensional.
    """

    def _fuerza_bruta_pequena(puntos: list) -> tuple:
        """Fuerza bruta para conjuntos pequeños (caso base)."""
        min_dist = float('inf')
        mejor_par = (None, None)
        n = len(puntos)
        for i in range(n):
            for j in range(i + 1, n):
                d = distancia_3d(puntos[i], puntos[j])
                if d < min_dist:
                    min_dist = d
                    mejor_par = (puntos[i], puntos[j])
        return min_dist, mejor_par

    def _franja_cercanos(franja: list, delta: float) -> tuple:
        """Busca el par más cercano en el slab (franja) central."""
        min_dist = delta
        mejor_par = (None, None)

        # Ordenar por coordenada Y
        franja.sort(key=lambda p: p[1])

        # Comparar cada punto con los siguientes mientras la diferencia en Y < delta
        for i in range(len(franja)):
            j = i + 1
            while j < len(franja) and (franja[j][1] - franja[i][1]) < min_dist:
                d = distancia_3d(franja[i], franja[j])
                if d < min_dist:
                    min_dist = d
                    mejor_par = (franja[i], franja[j])
                j += 1

        return min_dist, mejor_par

    def _dyv_recursivo(puntos_x: list) -> tuple:
        """Función recursiva principal."""
        n = len(puntos_x)

        # Caso base
        if n <= 3:
            return _fuerza_bruta_pequena(puntos_x)

        # Dividir por la mediana de X
        mitad = n // 2
        punto_medio = puntos_x[mitad]

        izquierda = puntos_x[:mitad]
        derecha = puntos_x[mitad:]

        # Conquistar
        dist_izq, par_izq = _dyv_recursivo(izquierda)
        dist_der, par_der = _dyv_recursivo(derecha)

        # Mínimo de ambas mitades
        if dist_izq < dist_der:
            delta = dist_izq
            mejor_par = par_izq
        else:
            delta = dist_der
            mejor_par = par_der

        # Combinar: crear el slab con puntos a distancia < delta del plano X = mediana
        franja = []
        for p in puntos_x:
            if abs(p[0] - punto_medio[0]) < delta:
                franja.append(p)

        dist_franja, par_franja = _franja_cercanos(franja, delta)

        if dist_franja < delta:
            return dist_franja, par_franja
        else:
            return delta, mejor_par

    # Ordenar por X
    puntos_ordenados = sorted(lista, key=lambda p: p[0])
    dist, par = _dyv_recursivo(puntos_ordenados)
    return dist, par[0], par[1]


# Ejecutamos y medimos el tiempo
inicio = time.time()
distancia, p1, p2 = puntos_cercanos_3d_dyv(LISTA_3D)
fin = time.time()

print(f"Divide y Vencerás 3D:")
print(f"  Puntos más cercanos: {p1} y {p2}")
print(f"  Distancia: {distancia:.4f}")
print(f"  Tiempo: {fin - inicio:.6f} segundos")
````

### Comparación de tiempos de ejecución

Ejecutamos todos los algoritmos y comparamos sus tiempos para ver la diferencia entre fuerza bruta O(n²) y Divide y Vencerás O(n log n / n log² n).

````python
print("=" * 70)
print("COMPARACIÓN DE TODOS LOS ALGORITMOS")
print("=" * 70)

# --- 1D ---
print("\n--- 1D (1000 puntos) ---")

inicio = time.time()
d1, pa1, pb1 = puntos_cercanos_1d_fuerza_bruta(LISTA_1D)
t_fb_1d = time.time() - inicio

inicio = time.time()
d2, pa2, pb2 = puntos_cercanos_1d_dyv(LISTA_1D)
t_dyv_1d = time.time() - inicio

print(f"  Fuerza Bruta:      dist={d1}, par=({pa1}, {pb1}), tiempo={t_fb_1d:.6f}s")
print(f"  Divide y Vencerás: dist={d2}, par=({pa2}, {pb2}), tiempo={t_dyv_1d:.6f}s")
if t_dyv_1d > 0:
    print(f"  Speedup: {t_fb_1d / t_dyv_1d:.1f}x más rápido con DyV")

# --- 2D ---
print("\n--- 2D (1000 puntos) ---")

inicio = time.time()
d3, pa3, pb3 = puntos_cercanos_2d_fuerza_bruta(LISTA_2D)
t_fb_2d = time.time() - inicio

inicio = time.time()
d4, pa4, pb4 = puntos_cercanos_2d_dyv(LISTA_2D)
t_dyv_2d = time.time() - inicio

print(f"  Fuerza Bruta:      dist={d3:.4f}, par={pa3} y {pb3}, tiempo={t_fb_2d:.6f}s")
print(f"  Divide y Vencerás: dist={d4:.4f}, par={pa4} y {pb4}, tiempo={t_dyv_2d:.6f}s")
if t_dyv_2d > 0:
    print(f"  Speedup: {t_fb_2d / t_dyv_2d:.1f}x más rápido con DyV")

# --- 3D ---
print("\n--- 3D (1000 puntos) ---")

inicio = time.time()
d5, pa5, pb5 = puntos_cercanos_3d_fuerza_bruta(LISTA_3D)
t_fb_3d = time.time() - inicio

inicio = time.time()
d6, pa6, pb6 = puntos_cercanos_3d_dyv(LISTA_3D)
t_dyv_3d = time.time() - inicio

print(f"  Fuerza Bruta:      dist={d5:.4f}, par={pa5} y {pb5}, tiempo={t_fb_3d:.6f}s")
print(f"  Divide y Vencerás: dist={d6:.4f}, par={pa6} y {pb6}, tiempo={t_dyv_3d:.6f}s")
if t_dyv_3d > 0:
    print(f"  Speedup: {t_fb_3d / t_dyv_3d:.1f}x más rápido con DyV")

# --- Resumen ---
print("\n" + "=" * 70)
print("RESUMEN DE COMPLEJIDADES")
print("=" * 70)
print(f"{'Algoritmo':<30} {'Complejidad':<20}")
print("-" * 50)
print(f"{'FB 1D, 2D, 3D':<30} {'O(n²)':<20}")
print(f"{'DyV 1D':<30} {'O(n log n)':<20}")
print(f"{'DyV 2D y 3D':<30} {'O(n log² n)':<20}")
````

### Conclusiones

1. **Fuerza Bruta** es el enfoque más sencillo de implementar pero tiene complejidad O(n²), lo que lo hace inviable para conjuntos de datos grandes.

2. **Divide y Vencerás** mejora significativamente el rendimiento:
   - En **1D**: Basta con ordenar y comparar vecinos → O(n log n)
   - En **2D** y **3D**: El algoritmo de Shamos & Hoey divide el espacio, resuelve recursivamente y combina con la franja/slab central → O(n log² n)

3. **Ambos métodos dan el mismo resultado** (misma distancia mínima y mismo par de puntos), lo que valida la corrección de las implementaciones.

4. **A medida que crece n**, la diferencia de tiempos entre fuerza bruta y DyV se hace cada vez más notable, ya que n² crece mucho más rápido que n log n.

5. La extensión de 2D a 3D es directa: solo cambia la fórmula de distancia y el hecho de que la "franja" pasa a ser un "slab" tridimensional, pero la estructura del algoritmo DyV se mantiene.
